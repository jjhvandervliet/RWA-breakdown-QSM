# Conclusions

In this project we have looked at the breakdown of the rotating wave approximation of the Jaynes cummimgs hamiltonian. Our main goal was to find out when this approximation fails by looking at many aspects of our system and seeing how the RWA causes deviation from the full JC hamiltonian including counter rotating terms. This difference between the hamiltonians can cause the breakdown in primarily the high regime. 

We analysed how the eigenenergies of the eigenstates changed for our system when there was an increase in coupling and found that RWA induces a significant error in the eigenenergies, when the coupling charcterised by the parameter g is increased. For strong coupling, g beyond 0.5 there is a error of more than 10%. This shows that RWA is only valid for the weak coupling regime. We can see that the squeezing effect, although preserved also is of much lower magnitude with the RWA. In addition, we also made the observation that central levels seem to deviate significantly over detuning at higher g, detuning itself at lower coupling does not play much of a role in the breakdown. 

We also found that when the coupling becomes stronger (Arround g $\approx$ 0.5) for the wigner quasi-probability function of the ground state breaks down. For the full Hamiltonian it evolves into a "cat" state while the RWA ground state develops multiple peaks without fringes as $g$ increases, failing to capture the superposition. Incorporating only noise into the system, it is revealed that the RWA breaks down easily for the average photon number even for low coupling strengths. Greater photon loss rates decreases the lower bound resulting in breakdown of the RWA at even lower values. The excited-state probability shows a structured sensitivity of the breakdown of the RWA at low to moderate values of the coupling strength. Modifying the noise parameters changes the dynamics of the relative deviation for both observables differently. Measurement back-action appears to fundamentally alter dynamics, resulting in the measurement protecting against the breakdown of the RWA for all values of the coupling strength at distinct time spans. Measurement rates effectively act analogously to delaying the dynamics that lead to the breakdown of the RWA. It can therefore be concluded that performing measurements could potentially benefit a system modelled by the Jaynes-Cummings Hamiltonian as it can possibly omit the RWA breakdown.
